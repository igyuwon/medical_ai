```python
# 사회 컬럼을 삭제, 음악 컬럼을 추가
# 음악의 점수는 과학 점수가 80점 이상이면 음악점수는 80점, 80점 미만은 65점 입력
# 합계 컬럼을 추가해서 국어+영어+수학+과학+음악
# 컬럼의 위치는 음악 과학 바로 뒤로 컬럼 위치 변경
df.drop(columns=['사회'],inplace=True)
df
```
