### drop - Nan처리 : 지우기

- axis : index,columns
- how : any,all

```python
# Nan 가 들어가 있는 컬럼을 전체 삭제
df.dropna(axis='columns',inplace=True)
```

```python
# Nan 들어가 있는 컬럼에서 모든 컬럼이 Nan 인 것만 삭제
df.dropna(axis='columns',how='all', inplace=True)
```

```python
# Nan 들어가 있는 컬럼에서 1개라 Nan이 포함되어 있으면 삭제
df.dropna(axis='columns',how='any',inplace=True)
```

```python
# Nan 들어가 있는 index 에서 1개라도 Nan이 포함되어 있으면 삭제
df.dropna(axis='index',how='any',inplace=True)
```

```python
df.dropna(axis='index',inplace=True) # 디폴트가 any
```
