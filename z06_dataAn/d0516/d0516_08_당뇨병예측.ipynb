#### 나이가 50이상인 사람은 몇명인가

```python
len(df[df['Age']>=50])
```

```python
# 나이가 50이상이면서 당뇨병 환자인 사람은 몇 명인지 출력하시오
filt = (df['Age']>=50) & (df['Outcome']==1)
len(df[filt])
```

```python
# 각 항목별 상관관계가 있는지 없는지 판단
df.corr()
```

```python
data_corr = df.corr()
fig = plt.figure(figsize=(7,7))
sns.heatmap(data_corr,annot=True)
plt.show()
```

```python
# 당뇨병 환자 원 그래프를 출력하시오.
values = df['Outcome'].value_counts().values
labels = ['비환자','환자']
```

```python
explode=[0,0.1]
plt.title('당뇨병 환자분포')
plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%', startangle=90) 
```
