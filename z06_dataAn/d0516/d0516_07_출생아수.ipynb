```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 표시 해결
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' # Mac사용자
matplotlib.rcParams['font.size'] = '10' # 글자크기
```

```python
df = pd.read_excel('../z20_data/stat_142801.xls',skiprows=2,nrows=2,index_col=0)
df
```

```python
# index에 특수문자 있는지 체크
df.index.values
```

```python
df.rename(index={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'},inplace=True)
```

```python
df.loc['출생아 수']
```

```python
df.columns
```

```python
df.iloc[0]
```

```python
df = df.T
```

```python
# 화면사이즈를 확대
plt.figure(figsize=(13,5))
x = df.index
y = df['출생아 수']
plt.bar(x,y,label='출생아 수',color='#ff812d')
plt.ylim(250,700)
plt.yticks([300,400,500,600])
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+3,txt,ha='center',color='black')
plt.legend()
plt.show()
```

```python
# 출생아 수는 막대그래프로
# 합계 출산
plt.figure(figsize=(13,5))
x = df.index
y_1 = df['합계 출산율']
plt.plot(x,y_1,'--o',c='#ffd100',linewidth=3)
plt.ylim(0,1.5)
plt.yticks([0,1])
for i, txt in enumerate(y_1):
    plt.text(x[i],y_1[i]+0.05,txt,ha='center',color='black')
plt.legend()
plt.show()



```

```python
fig, ax1 = plt.subplots(figsize=(10,4))
# 상단 타이틀 출력
fig.suptitle('출생아 수 및 합계 출산율')
ax1.bar(x,y,color='#ff812d')
ax1.set_ylabel('출생아 수')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i, txt in enumerate(y):
    ax1.text(x[i],y[i]+3,txt,ha='center',color='black')

ax2 = ax1.twinx()
ax2.plot(y_1,'-o',c='#ffd100',linewidth=3,markersize=10,markeredgecolor='w',markeredgewidth=3)
ax1.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
for i, txt in enumerate(y_1):
    ax2.text(x[i],y_1[i]+0.05,txt,ha='center',color='black')
    
    
plt.legend()
plt.show()
```
