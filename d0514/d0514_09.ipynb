```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 표시 해결
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
matplotlib.rcParams['font.size'] = '10' # 글자크기
```

```python
df = pd.read_csv('../z20_data/stu_score1.csv')
df.head(10)
```

```python
h_df = df.head(20)
x = h_df['id'].head(20)
y = h_df['total'].head(20)
y1 = h_df['kor'].head(20)
y2 = h_df['eng'].head(20)
y3 = h_df['math'].head(20)
```

```python
plt.bar(x,y)
plt.ylim(150,300)
plt.xticks(rotation=90)

plt.show()
```

```python
h_df.shape
h_df.shape[0]
```

```python
idx = np.arange(h_df.shape[0])
idx
```

```python
plt.figure(figsize=(15,5))
plt.bar(idx-0.3,y1,width=0.3) # 국어 점수
plt.bar(idx,y2,width=0.3) # 영어 점수

for i, txt in enumerate(y2):
    plt.text(idx[i],y2[i],txt,ha='center')

plt.bar(idx+0.3,y3,width=0.3) # 수학 점수

plt.xticks(idx,x,rotation=90)

plt.show()
```
