```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 표시 해결
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
matplotlib.rcParams['font.size'] = '10' # 글자크기
```

```python
df = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3,thousands = ',',usecols='B,E:Y',index_col='행정기관')

df1 = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3,thousands = ',',usecols='B,AY:BS',index_col='행정기관')

df2 = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3,thousands = ',',usecols='B,CS:DM',index_col='행정기관')

```

```python
df.loc['전국  ']=df.loc['전국  '].astype(int)
df1.loc['전국  ']=df1.loc['전국  '].astype(int)
df2.loc['전국  ']=df2.loc['전국  '].astype(int)
df
```

```python
df['0~14세'] = df.loc['전국  '][0:3].sum() # 1월
df1['0~14세.2'] = df1.loc['전국  '][0:3].sum() # 2월
df2['0~14세.4'] = df2.loc['전국  '][0:3].sum() # 3월
```

```python
# 15~64세
df['15~64세'] = df.loc['전국  '][3:13].sum()
df1['15~64세.2'] = df1.loc['전국  '][3:13].sum()
df2['15~64세.4'] = df2.loc['전국  '][3:13].sum()

# 65세이상
df['65세이상'] = df.loc['전국  '][13:21].sum()
df1['65세이상.2'] = df1.loc['전국  '][13:21].sum()
df2['65세이상.4'] = df2.loc['전국  '][13:21].sum()
```

```python
x = '2023년 1월'
y1 = df.loc['전국  ']['0~14세']
y2 = df.loc['전국  ']['15~64세']
y3 = df.loc['전국  ']['65세이상']
x1 = '2023년 2월'
y11 = df1.loc['전국  ']['0~14세.2']
y21 = df1.loc['전국  ']['15~64세.2']
y31 = df1.loc['전국  ']['65세이상.2']

x2 = '2023년 3월'
y12 = df2.loc['전국  ']['0~14세.4']
y22 = df2.loc['전국  ']['15~64세.4']
y32 = df2.loc['전국  ']['65세이상.4']

# 그래프 크기 설정
plt.figure(figsize=(8, 6))

# 막대 그래프 그리기
# 1월
plt.bar(x, y1, width=0.3, label='0~14세.1월', color='skyblue')
plt.bar(x, y2, width=0.3, bottom=y1, label='15~64세.1월', color='lightgreen')
plt.bar(x, y3, width=0.3, bottom=y1+y2, label='65세이상.1월', color='salmon')

# 2월
plt.bar(x1, y11, width=0.3, label='0~14세.2월', color='blue')
plt.bar(x1, y21, width=0.3, bottom=y11, label='15~64세.2월', color='green')
plt.bar(x1, y31, width=0.3, bottom=y11+y21, label='65세이상.2월', color='red')

# 3월
plt.bar(x2, y12, width=0.3, label='0~14세.3월', color='brown')
plt.bar(x2, y22, width=0.3, bottom=y11, label='15~64세.3월', color='silver')
plt.bar(x2, y32, width=0.3, bottom=y11+y21, label='65세이상.3월', color='gold')

# 레전드 설정
plt.legend(loc=(1.1,0.3), ncol=1)

# 레이블 추가
plt.text(x, y1 / 2, y1, ha='center', va='center', color='black')
plt.text(x, y1 + y2 / 2, y2, ha='center', va='center', color='black')
plt.text(x, y1 + y2 + y3 / 2, y3, ha='center', va='center', color='black')

plt.text(x1, y11 / 2, y11, ha='center', va='center', color='black')
plt.text(x1, y11 + y21 / 2, y21, ha='center', va='center', color='black')
plt.text(x1, y11 + y21 + y31 / 2, y31, ha='center', va='center', color='black')

plt.text(x2, y12 / 2, y12, ha='center', va='center', color='black')
plt.text(x2, y12 + y22 / 2, y22, ha='center', va='center', color='black')
plt.text(x2, y12 + y22 + y32 / 2, y32, ha='center', va='center', color='black')

# 제목 및 축 레이블 설정
plt.title('2023년 인구 구성')
plt.xlabel('기간')
plt.ylabel('인구수')

# y축 범위 설정
plt.ylim(0, y1 + y2 + y3 + 50)

# 그래프 표시
plt.show()
```
