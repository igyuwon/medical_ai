```python
sizes = np.random.rand(8)*300
# s는 산점도 그래프 점의 크기 조절
plt.scatter(df['영어'],df['수학'],s=sizes)
```

```python
# 학년별 크기 조절
sizes = df['학년']*150
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
# 학년별 색상 변경
sizes = df['학년']*150
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis')
plt.xlabel('영어점수')
plt.ylabel('수학점수')
# 컬러바 눈금표시 : ticks, 라벨 : label, 크기조절 : shrink, orientation : 하단배치
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5, orientation='horizontal')
```
